```python
#Task 2.16
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]

night_1=set(night1)
night_2=set(night2)

# 1
common_values=set.intersection(night_1, night_2)
print(f"Common Values:{common_values}")

# 2
differences=night_1.difference(night_2), night_1-night_2
print(f"Differences: {differences}")

# 3 
new_night1=[]
for n in night1:
    if n not in new_night1:
     new_night1.append(n)
print(new_night1)
        
new_night2=[]
for n in night2:
    if n not in new_night2:
     new_night2.append(n)
     
print(new_night2)


# 4
night_1_new = [n for n in night_1 if n in common_values]
print(f"Updated list:{night_1_new}")
```
