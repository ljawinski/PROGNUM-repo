```python
# Task 4.12
import math
from matplotlib import pyplot as plt


# creates complex rotor
angle_degs = 30
theta = math.radians(angle_degs)

rotor = math.cos(theta) + 1j*math.sin(theta)
print(rotor, abs(rotor)) 

# creates plot of a square
x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])

plt.plot(x,y)

# rotates the square
x1=rotor*(x+y*1j)

# plots the rotated square
plt.plot(x1.real, x1.imag, color='red')
```

```python
# Task 4.16
import scipy
from scipy import integrate
from astropy.cosmology import Planck18
import math

# defines constants
H0=Planck18.H0.value
Om0=Planck18.Om0
Ode0=Planck18.Ode0

# defines function 
def function(z): 
  Hz=H0*math.sqrt(Om0*((1+z)**3)+Ode0)
  f=1/((1+z)*Hz)
  return f
   
# calculates the integral    
z=float(input("value for the redshift: "))
a=0
b=z
tLz=integrate.quad(function, a, b)
print(tLz)

# converts km/Mpc*s to seconds
A=tLz[0]
seconds=A*3.08568e19

# converts seconds to Gyrs
t=seconds/3.156e16
print(t)
```

```python
# Task 4.21
import numpy as np 
from matplotlib import pyplot as plt
from scipy.stats import norm

# generates random data
X=np.random.normal(loc=0, scale=1, size=100000)
Y=np.random.normal(loc=2, scale=1, size=100000)

# plots the data in a histogram
plt.hist(X, color='blue', bins=30, density=True, label="X", alpha=0.5)
plt.hist(Y, color='red', bins=30, density=True, label="Y", alpha=0.5)

# adds the probability distribution graph
x_vals=np.linspace(-10,10,100000)
pdf_x=norm.pdf(x_vals, 0, 1)
pdf_y=norm.pdf(x_vals,2,1)
plt.plot(x_vals, pdf_x, 'k:', linewidth=3, label='X (PDF)')
plt.plot(x_vals, pdf_y, 'k--', linewidth=3, label='Y (PDF)')

# adds labels to the axes and a title
plt.xlabel("Values")
plt.ylabel("Probability Density")
plt.title("1D Histogram + PDF")

# adds gridlines
plt.grid(True ,linestyle='--', alpha=0.6)

# adds distribution labels
plt.annotate(r'$\mathcal{N}(2,1)$',
             xy=(-3, 0.15),
             xytext=(-8,0.15),
             color='blue',
             fontsize=14,
             arrowprops=dict(arrowstyle="-|>", color='blue', lw=5))

plt.annotate(r'$\mathcal{N}(2,1)$',
             xy=(5, 0.15),
             xytext=(7,0.15),
             color='red',
             fontsize=14,
             arrowprops=dict(arrowstyle="-|>", color='red', lw=5))


# adds a legend
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# example data
x=np.random.normal(loc=0, scale=1, size=100000)
y=np.random.normal(loc=2, scale=1, size=100000)

# compute histogram
H, xedges, yedges = np.histogram2d(x, y, bins=(30,30))

# convert to probability
H = H / H.sum()

# flatten and sort densities
Hflat = H.flatten()
Hsorted = np.sort(Hflat)[::-1]
cumsum = np.cumsum(Hsorted)

# confidence levels
probs = [0.683, 0.954, 0.997]

levels = []
for p in probs:
    levels.append(Hsorted[cumsum >= p][0])
    
# sort levels for contour
levels = np.sort(np.unique(levels))

# bin centers
xcenters = (xedges[:-1] + xedges[1:]) / 2
ycenters = (yedges[:-1] + yedges[1:]) / 2
X, Y = np.meshgrid(xcenters, ycenters)

# plot histogram
plt.imshow(
    H.T,
    origin="lower",
    extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]],
    cmap="Blues",
    aspect="auto"
)

plt.colorbar()

# adds contour lines and labels them 
CS = plt.contour(X, Y, H.T, levels=levels, colors="red", linewidths=1)

fmt = {
    levels[0]: "3σ",
    levels[1]: "2σ",
    levels[2]: "1σ"
}

plt.clabel(CS, fmt=fmt)

# adds axes labels
plt.xlabel("x")
plt.ylabel("y")

# adds grid
plt.grid(linestyle='--', color="black")

# adds title
plt.title("2d Histogram + Contour Lines")
plt.show()

```
