```python
# Task 3.10
from math import pi

# defines function
def cylinder(radius, height):
    r=radius
    h=height
    A=2*pi*r**2+2*pi*h
    V=pi*r**2*h
    return A, V

# calculates ratio
Area, Volume = cylinder(1,2)
ratio=Area/Volume
print(ratio)
    
```

```python
# Task 3.17
import numpy as np
import numpy.random
import math

test_array=numpy.random.normal(loc=0.0, scale= 5, size=100)

# creates function to calculate 
def stdev():
    test_array=numpy.random.normal(loc=0.0, scale= 5, size=100)
    m=1/100*sum(test_array)
    sigma=math.sqrt(1/100*sum((test_array-m)**2))
    return(sigma)
    
    
a=numpy.std(test_array)
b=stdev()
print(a)
print(b)
```

```python
# Task 3.4
import math
import numpy as np

# creates list
data = [2.3, 3.5, 5.6, 12.8, 10.3, 7.2, 5.0, 2.1, -2.0]

# delete negative values
for n in data:
    if n<0:
        data.remove(n)
        print(data)

# calculates average intensity
m=0
for x in data:
    m=m+x

mean = (1/8)*m
print(mean)

# calculates standard deviation
s=0
for y in data:
    s=s+(y-mean)**2

standard_deviation=math.sqrt((1/8)*(s))
print(standard_deviation)

# converts data to a numpy array
data2=np.array(data)

# calculates standard deviation using a numpy function
standard_deviation2=np.std(data2)
print(standard_deviation2)



```

```python
# Task 3.6
import math
from scipy.constants import pi
from scipy.special import erf

# defines function that calculates the integral
def function(x):
    total=0
    factorial=1
    for k in range(100):
        if k>0:
            factorial*=k
        numerator=((-1)**k)*(x**(2*k+1))
        denominator=(2*k+1)*factorial
        s=numerator/denominator
        total=total+s
    erfx=(2/math.sqrt(math.pi))*total
    return erfx

# calls function for x=1
print(function(1))

# calculates the value using a scipy function
print(erf(1))
```

```python
# Task 3.11
import math

# defines a function that determines wether a,b,c is a pythagorean triples
def istriple(a, b, c):
    if a**2+b**2==c**2:
     return(True)
     print("True")
    else:
     return(False)
     print("False")

# calls the function for some test values
istriple(3, 4, 5)

# gives first ten values with the given condition
x=[]
for a in range(0,100):
    for b in range(a,100):
        for c in range(0,100):
            if istriple(a,b,c)==True:
                x.append((a,b,c))

print(x[:10])





    
```

```python
# Task 3.13

# defines function
def fibonacci(N,M):
    #generates list with numbers in the fibonacci sequence
    a=0
    b=1
    fibonacci=[]
    for x in range(110):
        fibonacci.append(a)
        a, b=b, a+b
    # filters numbers with the given conditions
    numbers=[]
    for i in fibonacci:
     y=i/M
     if i<N and y - int(y)==0.0:
         numbers.append(i)
    print(numbers)
    
# calls function for given values
fibonacci(100,7)

        
    
```

```python
# Task 3.23
from matplotlib import pyplot as plt

distance =[18.49, 35.48, 35.48, 11.34, 7.12, 9.82, 23.66, 44, 66.7, 52.4, 55, 25.6, 26.19]
velocity = [2094.75, 2320.87, 1224.9, 872.13, 432.2, 751, 14.10, 27.00, 5292, 2550, 5253, 2419, 2158]
distance.sort()
velocity.sort()

# creates plot of the given data
plt.plot(distance, velocity, marker='*', color='blue', label="measured data")


# generates data expected due to hubbles law
velocity_expected=[]
for x in distance:
    h=67.66*x
    velocity_expected.append(h)
print(velocity_expected)

plt.plot(distance, velocity_expected, color='red', label="expected data")
plt.title(r"$Plot\ of\ the\ Lemaitre-Hubble\ relation$")
plt.xlabel(r"$Distance\ in\ Mpc$")
plt.ylabel(r"$Velocity\ in\ km/s$")
plt.legend()
plt.show

```

```python
# Task 3.24
import numpy as np
import math

distance =[18.49, 35.48, 35.48, 11.34, 7.12, 9.82, 23.66, 44, 66.7, 52.4, 55, 25.6, 26.19]
velocity = [2094.75, 2320.87, 1224.9, 872.13, 432.2, 751, 14.10, 27.00, 5292, 2550, 5253, 2419, 2158]

# creates function that calculates the pearson correlation
def pearson(X, Y):
 x=np.array(X)
 y=np.array(Y)
 x_mean=np.mean(x)
 y_mean=np.mean(y)
                
 numerator=np.sum((x-x_mean)*(y-y_mean))
 denominator=math.sqrt((np.sum(x-x_mean)**2))*np.sum((y-y_mean)**2)
 rxy=numerator/denominator
    
 print("Pearson correlation coefficient: " ,rxy)
 return rxy
    
# calls function for given data
pearson(distance, velocity)

# since rxy>1 we can conclude that there is a very strong linear correlation between distance and velocity




```

```python
# 3.25
import math
import numpy as np
from matplotlib import pyplot as plt

distance=[18.49, 35.48, 35.48, 11.34, 7.12, 9.82, 23.66, 44, 66.7, 52.4, 55, 25.6, 26.19]
velocity=[2094.94, 2320.87, 1224.9, 872.13, 432.2, 751, 1410, 2700, 5292, 2550, 5253, 2419, 2158]

distance.sort()
velocity.sort()

x=np.array(distance)
y=np.array(velocity)

x_mean=np.mean(x)
y_mean=np.mean(y)

Sxy=np.sum(x*y)
Sxx=np.sum(x**2)

H0=Sxy/Sxx
H_lit=67.66

velocity_best_line=[]
velocity_hubble_constant=[]

for a in distance:
    v1=H0*a
    velocity_best_line.append(v1)

    v2=H_lit*a
    velocity_hubble_constant.append(v2)

plt.plot(distance, velocity, label=r"$measured\ data$")
plt.plot(distance, velocity_best_line, label=r"$linear\ fit\ through\ data$")
plt.plot(distance, velocity_hubble_constant, label=r"$expect$ed\ due\ to\ Hubble\ constant$")
plt.legend()
plt.title(r"$Plot\ of\ Hubble\ data$")
plt.xlabel(r"$distance\ in\ Mpc$")
plt.ylabel(r"$velocity\ in\ km/s$")
plt.show
```
